This notebook demonstrates basic NLP preprocessing techniques including:
- Lowercasing
- Removing punctuation
- Removing stopwords
- Tokenization
- Stemming
- Lemmatization


```python
import nltk
import re
import string

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

## Sample Text
We will use a simple paragraph for preprocessing.

```python
text = """Natural Language Processing (NLP) is a field of artificial intelligence 
that focuses on understanding human language. Preprocessing is a crucial step 
in NLP pipelines!"""

text
```

## 1. Lowercasing the Text
Converts all characters into lowercase.

```python
lower_text = text.lower()
lower_text
```

## 2. Removing Punctuation

```python
clean_text = re.sub(f"[{string.punctuation}]", "", lower_text)
clean_text
```

## 3. Tokenization
Splitting text into individual words.

```python
tokens = nltk.word_tokenize(clean_text)
tokens
```

## 4. Removing Stopwords
Common words such as is, the, a are removed.

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]
filtered_tokens
```

## 5. Stemming
Using PorterStemmer to reduce words to their root forms.

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
stemmed_words
```

## 6. Lemmatization
More accurate root word extraction than stemming.

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
lemmatized_words
```

## Final Output
Comparison summary.

```python
print("Original Text:\n", text)
print("\nTokens:\n", tokens)
print("\nAfter Stopword Removal:\n", filtered_tokens)
print("\nStemmed Words:\n", stemmed_words)
print("\nLemmatized Words:\n", lemmatized_words)
```
